# 1.6 Regression and Gradient Descent

In this problem, we will look at OLS and the gradient descent algorithm.

**Part (a):** Read in the synthetic data matrix `syn_X.csv` and the vector `syn_y.csv` of “observations”.

```python
syn_X = pd.read_csv("./PSet1_Data/syn_X.csv", header=None)
syn_y = pd.read_csv("./PSet1_Data/syn_y.csv", header=None)
```

Compute the OLS estimator $\hat{\beta}$ by matrix inversion.

```python
y = np.array(syn_y[0])
X = np.transpose(np.array([syn_X[0], syn_X[1]]))
betahat = np.linalg.solve( np.transpose(X) @ X, np.transpose(X) @ y)
betahat
```

**Part (b):** Implement the gradient descent for the least squares problem.

*Solution: We aim to minimise $f(\vec{\beta}) = \| \vec{y} - X \vec{\beta} \|^2$. Notice $\vec{\nabla}f = 2(X^T X \vec{\beta} - X^T \vec{y})$.*

```python
def ols_by_descent(X, y, beta0, alpha=0.1, n=50, print_all=False):

    beta = beta0
    
    if print_all==True:
        beta_list = [0 for i in range(n)]
        beta_list[0] = beta0
        for i in range(1,n):
            old_beta = beta
            beta = old_beta - 2*alpha*((np.transpose(X) @ X) @ old_beta - np.transpose(X) @ y)
            beta_list[i] = beta
        return beta_list
    
    for i in range(1,n):
        old_beta = beta
        beta = old_beta - 2*alpha*(np.transpose(X) @ (X @ old_beta) - np.transpose(X) @ y)
    return beta
```

Run it on our synthetic testdata.  As a function of the iteration $t$, plot the mean squared error and the distance $\|\vec{\beta}^{(t)} - \hat{\beta} \|$ of your current iterates (in separate plots).  Play with different initializations $\vec{\beta}^0$ and stepsizes. What do you observe? Explain.  What would be an optimal step size?

*Solution: First implement with varying $\alpha$.*

```python
beta_vec1 = ols_by_descent(X,y,np.array([1.5,-4.5]), alpha=7e-3, n=100, print_all=True)
beta_vec2 = ols_by_descent(X,y,np.array([1.5,-4.5]), alpha=6e-3, n=100, print_all=True)
beta_vec3 = ols_by_descent(X,y,np.array([1.5,-4.5]), alpha=1e-3, n=100, print_all=True)
beta_vec4 = ols_by_descent(X,y,np.array([1.5,-4.5]), alpha=5e-4, n=100, print_all=True)
err_vec1 = [np.linalg.norm(beta_vec1[i] - betahat) for i in range(len(beta_vec1))]
err_vec2 = [np.linalg.norm(beta_vec2[i] - betahat) for i in range(len(beta_vec2))]
err_vec3 = [np.linalg.norm(beta_vec3[i] - betahat) for i in range(len(beta_vec3))]
err_vec4 = [np.linalg.norm(beta_vec4[i] - betahat) for i in range(len(beta_vec4))]
```

```python
plt.plot(err_vec1)
plt.plot(err_vec2)
plt.plot(err_vec3)
plt.plot(err_vec4)
plt.title("Distance between betahat and beta's in gradient descent (alpha changes)")
plt.legend(["alpha=7e-3", "alpha=6e-3", "alpha=1e-3", "alpha=5e-4"])
```

*Observation: When $\alpha$ is sufficiently small, the algorithm converges. Of course, when $\alpha$ is not too small, the algorithm converges quicker. The optimal $\alpha$ seems to be around $0.006$. Notice when $\alpha$ is too large, overshooting occurs and the algorithm diverges as a result.*

```python
beta_vec0 = ols_by_descent(X,y,np.array([1.5,-4.5]), alpha=7.5e-3, n=100, print_all=True)
err_vec0 = [np.linalg.norm(beta_vec0[i] - betahat) for i in range(len(beta_vec0))]
plt.plot(err_vec0)
plt.title("Distance between betahat and beta's in gradient descent (alpha changes)")
plt.legend(["alpha=0.0075"])
```

*One can then use this $\alpha$ and observe the implementation with varying $\beta_0$.*

```python
beta_vec5 = ols_by_descent(X,y,np.array([0,0]), alpha=6e-3, n=100, print_all=True)
beta_vec6 = ols_by_descent(X,y,np.array([5,5]), alpha=6e-3, n=100, print_all=True)
beta_vec7 = ols_by_descent(X,y,np.array([-10,-10]), alpha=6e-3, n=100, print_all=True)
err_vec5 = [np.linalg.norm(beta_vec5[i] - betahat) for i in range(len(beta_vec5))]
err_vec6 = [np.linalg.norm(beta_vec6[i] - betahat) for i in range(len(beta_vec6))]
err_vec7 = [np.linalg.norm(beta_vec7[i] - betahat) for i in range(len(beta_vec7))]
```

```python
plt.plot(err_vec5)
plt.plot(err_vec6)
plt.plot(err_vec7)
plt.title("Distance between betahat and beta's in gradient descent (beta0 changes)")
plt.legend(["beta0=[0,0]", "beta0=[5,5]", "beta0=[-5,-5]"])
```

*Observation: The performance doesn't change when initial value has changed. This is because our loss function has only a minimum, so the algorithm must converge to that global minimum. Moreover, the iteration required for convergent is approximately the same, this is because the values of gradients are adapted with the choice of initial data.*
